# Análise da curva de aprendizado para 10x10 um único agente

```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('data/statistics_10_10_1.csv')
sns.lineplot(data=df, x="episode", y=df['rewards'].ewm(span=100).mean()).set(title="Reward por episódios (valor médio considerando janela de 100 episódios)")
```
